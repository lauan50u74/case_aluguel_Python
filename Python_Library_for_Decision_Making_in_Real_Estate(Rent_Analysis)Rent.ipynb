```python
import pandas as pd
aluguel_casa = pd.read_excel('/houses_rent.xlsx')
```

```python
agg_city_price = aluguel_casa.groupby(["city"])["total"].agg("mean").reset_index()
agg_city_price
```

```python
agg_bathrooms_rooms_price = aluguel_casa.groupby(["bathroom","rooms"])["total"].agg("mean").reset_index()
agg_bathrooms_rooms_price
```

```python
import seaborn as sns
ax = sns.barplot(data=agg_bathrooms_rooms_price, x="rooms", y="total", color = 'yellow');
ax.set(xlabel = "qtd quartos",ylabel = "mdia de preo" );
```

```python
ax = sns.barplot(data=agg_bathrooms_rooms_price, x="bathroom", y="total", color = 'red');
ax.set(xlabel = "qtd banheiros",ylabel = "mdia de preo" );
```
